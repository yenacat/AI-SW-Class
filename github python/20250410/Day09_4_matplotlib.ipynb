```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-20, 21)  # -20부터 20까지의 정수로 이루어진 배열 생성
y1 = 2 * x  # f(x) = 2x
y2 = (1 / 3) * x**2 + 5  # f(x) = (1/3)x^2 + 5
y3 = -(x**2) - 5  # f(x) = -x^2 - 5

plt.plot(
    x, y1, "g--", x, y2, "r^-", x, y3, "b*:"
)  # 초록색 점선, 빨간색 삼각형, 파란색 별표로 함수를 표현
plt.axis([-30, 30, -30, 30])
plt.show()
```
