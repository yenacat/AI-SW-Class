```python
import numpy as np
import matplotlib.pyplot as plt


def mandelbrot(h, w, maxit=20):
    y, x = np.ogrid[-1.4 : 1.4 : h * 1j, -2 : 0.8 : w * 1j]
    c = x + y * 1j
    z = c
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + c
        diverge = z * np.conj(z) > 2**2
        div_now = diverge & (divtime == maxit)
        divtime[div_now] = i
        z[diverge] = 2

    return divtime


plt.imshow(mandelbrot(400, 400))
plt.show()
```

```python
# Q1.
# 1부터 25까지의 숫자로 이루어진 배열을 생성하세요.
# 배열의 각값을 0과 1 사이로 졍규화(normalization)하세요.
# 정규화란 주로 0에서 1사이의 범위로 데이터를 변환하는 작업
# 정규화 공식: (x - min) / (max - min)

import numpy as np
import matplotlib.pyplot as plt

question1_1 = np.arange(1, 26)
print(question1_1)

question1_2 = (question1_1 - np.min(question1_1)) / (
    np.max(question1_1) - np.min(question1_1)
)
print(question1_2)
```

```python
# Q2.
# 5 * 5 크기의 배열을 생성하고 배열의 값은 0에서 50 사이의 랜덤 정수로 구성됨.
# 생성한 배열에서 30보다 큰 값만 필터링하여 새 배열을 만드세요.
# 필터링 된 배열의 값 중 최댓값, 최솟값, 평균값을 계산하세요.

question2_1 = np.random.randint(0, 51, (5, 5))
print(question2_1)

question2_2 = question2_1[question2_1 > 30]
print(question2_2)

print(f"필터링 된 배열의 최댓값: {np.max(question2_2)}")
print(f"필터링 된 배열의 최솟값: {np.min(question2_2)}")
print(f"필터링 된 배열의 평균값: {np.mean(question2_2):.2f}")
```

```python
# Q3.
# 숫자 1부터 50까지의 배열을 생성하세요.
# 배열에서 짝수만 추출해 새로운 배열을 만드세요.
# 원본 배열과 추출한 배열을 막대 그래프로 시각화하세요.

question3_1 = np.arange(1, 51)
print(question3_1)

question3_even = question3_1[question3_1 % 2 == 0]
print(question3_even)

plt.bar(
    range(len(question3_1)), question3_1, color="blue", alpha=0.5, label="원본 배열"
)  # range 함수로 x축 값 생성, y축 값은 question3_1 배열 사용
plt.bar(
    range(len(question3_even)),
    question3_even,
    color="orange",
    alpha=0.5,
    label="추출한 배열",
)  # range 함수로 x축 값 생성, y축 값은 question3_even 배열 사용

plt.xlabel("데이터 순서 (Index)")  # X축 레이블 설정
plt.ylabel("값 (Value)")  # Y축 레이블 설정

plt.title("배열 데이터 시각화 (순서 기준)")  # 그래프 제목 추가
plt.legend()  # 범례 표시 (label 인수가 사용되었으므로)
plt.grid(axis="y", linestyle="--", alpha=0.7)  # y축 그리드 추가
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 숫자 1부터 50까지의 배열 생성
question3_1 = np.arange(1, 51)
print("원본 배열:")
print(question3_1)

# 2. 배열에서 짝수만 추출해 새로운 배열 생성
question3_even = question3_1[question3_1 % 2 == 0]
print("\n짝수 배열:")
print(question3_even)


# 3. 원본 배열과 추출한 배열을 막대 그래프로 시각화

# 하나의 Figure 안에 두 개의 Subplot (Axes) 생성 (1행 2열)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))  # 전체 그림 크기 조절

# 첫 번째 Subplot: 원본 배열 시각화
axes[0].bar(
    question3_1, question3_1, color="skyblue", width=0.8
)  # x축에 실제 숫자 사용, 색상 및 너비 지정
axes[0].set_title("원본 배열 (1-50)")
axes[0].set_xlabel("숫자")
axes[0].set_ylabel("값")
axes[0].grid(axis="y", linestyle="--", alpha=0.7)  # y축 그리드 추가

# 두 번째 Subplot: 짝수 배열 시각화
axes[1].bar(
    question3_even, question3_even, color="lightgreen", width=0.8
)  # x축에 실제 짝수 사용
axes[1].set_title("짝수만 추출한 배열")
axes[1].set_xlabel("숫자 (짝수)")
axes[1].set_ylabel("값")
axes[1].grid(axis="y", linestyle="--", alpha=0.7)  # y축 그리드 추가

# Subplot 간 간격 자동 조절
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

question3_1 = np.arange(1, 51)
question3_even = question3_1[question3_1 % 2 == 0]

# --- 스템 플롯 시각화 ---
fig, axes = plt.subplots(2, 1, figsize=(12, 8))  # 2행 1열 서브플롯

# 원본 배열 스템 플롯
markerline, stemlines, baseline = axes[0].stem(
    question3_1, question3_1, linefmt="grey", markerfmt="o", basefmt="black"
)
plt.setp(markerline, "color", "blue")  # 마커 색상 변경
axes[0].set_title("Stem Plot of Original Array (1-50)")
axes[0].set_xlabel("Number")
axes[0].set_ylabel("Value")
axes[0].grid(axis="y", linestyle=":")

# 짝수 배열 스템 플롯
markerline, stemlines, baseline = axes[1].stem(
    question3_even, question3_even, linefmt="grey", markerfmt="x", basefmt="black"
)
plt.setp(markerline, "color", "red")  # 마커 색상 변경
axes[1].set_title("Stem Plot of Even Numbers Array")
axes[1].set_xlabel("Even Number")
axes[1].set_ylabel("Value")
axes[1].grid(axis="y", linestyle=":")

plt.tight_layout()
plt.show()
```
