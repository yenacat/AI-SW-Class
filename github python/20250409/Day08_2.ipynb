```python
import numpy as np
import matplotlib.pyplot as plt


def mandelbrot(h, w, maxit=20):
    y, x = np.ogrid[-1.4 : 1.4 : h * 1j, -2 : 0.8 : w * 1j]
    c = x + y * 1j
    z = c
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + c
        diverge = z * np.conj(z) > 2**2
        div_now = diverge & (divtime == maxit)
        divtime[div_now] = i
        z[diverge] = 2

    return divtime


plt.imshow(mandelbrot(400, 400))
plt.show()
```

```python
# Q1.
# 1 25    .
#   0 1  (normalization).
#   0 1    
#  : (x - min) / (max - min)

import numpy as np
import matplotlib.pyplot as plt

question1_1 = np.arange(1, 26)
print(question1_1)

question1_2 = (question1_1 - np.min(question1_1)) / (
    np.max(question1_1) - np.min(question1_1)
)
print(question1_2)
```

```python
# Q2.
# 5 * 5      0 50    .
#   30      .
#      , ,  .

question2_1 = np.random.randint(0, 51, (5, 5))
print(question2_1)

question2_2 = question2_1[question2_1 > 30]
print(question2_2)

print(f"   : {np.max(question2_2)}")
print(f"   : {np.min(question2_2)}")
print(f"   : {np.mean(question2_2):.2f}")
```

```python
# Q3.
#  1 50  .
#      .
#       .

question3_1 = np.arange(1, 51)
print(question3_1)

question3_even = question3_1[question3_1 % 2 == 0]
print(question3_even)

plt.bar(
    range(len(question3_1)), question3_1, color="blue", alpha=0.5, label=" "
)  # range  x  , y  question3_1  
plt.bar(
    range(len(question3_even)),
    question3_even,
    color="orange",
    alpha=0.5,
    label=" ",
)  # range  x  , y  question3_even  

plt.xlabel("  (Index)")  # X  
plt.ylabel(" (Value)")  # Y  

plt.title("   ( )")  #   
plt.legend()  #   (label  )
plt.grid(axis="y", linestyle="--", alpha=0.7)  # y  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1.  1 50  
question3_1 = np.arange(1, 51)
print(" :")
print(question3_1)

# 2.      
question3_even = question3_1[question3_1 % 2 == 0]
print("\n :")
print(question3_even)


# 3.       

#  Figure    Subplot (Axes)  (1 2)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))  #    

#   Subplot:   
axes[0].bar(
    question3_1, question3_1, color="skyblue", width=0.8
)  # x   ,    
axes[0].set_title("  (1-50)")
axes[0].set_xlabel("")
axes[0].set_ylabel("")
axes[0].grid(axis="y", linestyle="--", alpha=0.7)  # y  

#   Subplot:   
axes[1].bar(
    question3_even, question3_even, color="lightgreen", width=0.8
)  # x   
axes[1].set_title("  ")
axes[1].set_xlabel(" ()")
axes[1].set_ylabel("")
axes[1].grid(axis="y", linestyle="--", alpha=0.7)  # y  

# Subplot    
plt.tight_layout()

#  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

question3_1 = np.arange(1, 51)
question3_even = question3_1[question3_1 % 2 == 0]

# ---    ---
fig, axes = plt.subplots(2, 1, figsize=(12, 8))  # 2 1 

#    
markerline, stemlines, baseline = axes[0].stem(
    question3_1, question3_1, linefmt="grey", markerfmt="o", basefmt="black"
)
plt.setp(markerline, "color", "blue")  #   
axes[0].set_title("Stem Plot of Original Array (1-50)")
axes[0].set_xlabel("Number")
axes[0].set_ylabel("Value")
axes[0].grid(axis="y", linestyle=":")

#    
markerline, stemlines, baseline = axes[1].stem(
    question3_even, question3_even, linefmt="grey", markerfmt="x", basefmt="black"
)
plt.setp(markerline, "color", "red")  #   
axes[1].set_title("Stem Plot of Even Numbers Array")
axes[1].set_xlabel("Even Number")
axes[1].set_ylabel("Value")
axes[1].grid(axis="y", linestyle=":")

plt.tight_layout()
plt.show()
```
