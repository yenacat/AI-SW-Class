```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

weather = pd.read_csv(weather_file, encoding="CP949")
weather["month"] = pd.DatetimeIndex(weather["일시"]).month

monthly = [None for x in range(12)]  # 달 별로 구분된 12개의 None 리스트 생성
yearly_temp = [0 for x in range(12)]  # 달 별 평균 기온을을 저장할 리스트 생성

# 월 별 평균 기온온을 구하기 위해 반복문을 돌려서 월별 데이터프레임을 생성

for i in range(12):
    monthly[i] = weather[weather["month"] == i + 1]  # 달 별로 분리
    yearly_temp[i] = monthly[i]["평균기온"].mean()

years = np.arange(1, 13)  # 1~12까지의 월을 생성
plt.bar(years, yearly_temp, color="green")
plt.xlabel("Month")
plt.ylabel("Temperature")

plt.show()
```

```python
df_1 = pd.DataFrame(
    {
        "A": ["a10", "a11", "a12"],
        "B": ["b10", "b11", "b12"],
        "C": ["c10", "c11", "c12"],
    },
    index=["가", "나", "다"],
)

df_2 = pd.DataFrame(
    {
        "B": ["b23", "b24", "b25"],
        "C": ["c23", "c24", "c25"],
        "D": ["d23", "d24", "d25"],
    },
    index=["다", "라", "마"],
)
```

```python
df_3 = pd.concat([df_1, df_2])

display(df_3)
```

```python
df_4 = pd.concat([df_1, df_2], join="inner")

display(df_4)
```

```python
# left outer

df_1.merge(df_2, how="left", on="B")
```

```python
# right outer

df_1.merge(df_2, how="right", on="B")
```

```python
# full outer

df_1.merge(df_2, how="outer", on="B")
```

```python
# inner

df_1.merge(df_2, how="inner", on="B")
```

```python
df_3 = df_1.merge(df_2, how="outer", left_index=True, right_index=True)

display(df_3)
```
