```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi * 2, 100)

fig = plt.figure()

plt.plot(x, np.sin(x), "r-")  # sin curve
plt.plot(x, np.cos(x), "b:")  # cos curve

fig.savefig("sin_cos_fig.png")  # save figure

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

triXAxis = np.linspace(-4, 4, 100)
sinYAxis = np.sin(triXAxis)
cosYAxis = np.cos(triXAxis)

plt.rcParams["font.family"] = "Malgun Gothic"  # 한글 폰트 설정
plt.rcParams["axes.unicode_minus"] = False  # 마이너스 기호 깨짐 방지

plt.title("삼각함수")
plt.plot(triXAxis, sinYAxis, "r-", label="sin curve")
plt.plot(triXAxis, cosYAxis, "b:", label="cos curve")
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([-1, -0.5, 0, 0.5, 1])


plt.xlabel("x 값")
plt.ylabel("y 값")
plt.legend(loc="best", fontsize=10, shadow=True, fancybox=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(211).set(xticks=[-4, -2, 0, 2, 4], yticks=[-1, -0.5, 0, 0.5, 1])

triXAxis = np.linspace(-4, 4, 100)
sinYAxis = np.sin(triXAxis)
cosYAxis = np.cos(triXAxis)

plt.plot(triXAxis, sinYAxis)
plt.plot(triXAxis, cosYAxis)
plt.title("Trigonometric functions")

plt.subplot(212).set(xlim=[0, 8], ylim=[0, 3])
logXAxis = np.linspace(1, 10, 100)
logYAxis = np.log(logXAxis)

plt.plot(logXAxis, logYAxis)
plt.title("Logarithmic functions")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("default")

fig, ax = plt.subplots(2, 2)  # 2행 2열의 서브플롯 생성

X = np.random.randn(100)  # 정규분포를 따르는 랜덤 숫자 100개 생성
Y = np.random.randn(100)
ax[0, 0].scatter(X, Y)  # 첫 번째 서브플롯에 산점도 그리기

X = np.arange(10)  # 0부터 9까지의 정수 배열 생성
Y = np.random.uniform(1, 10, 10)  # 1부터 10까지의 랜덤 숫자 10개 생성
ax[0, 1].bar(X, Y)  # 두 번째 서브플롯에 막대그래프 그리기

X = np.linspace(0, 10, 100)  # 0부터 10까지 100개의 점 생성
Y = np.cos(X)  # X에 대한 코사인 값
ax[1, 0].plot(X, Y)  # 세 번째 서브플롯에 선 그래프 그리기

Z = np.random.uniform(0, 1, (5, 5))  # 5x5 랜덤 배열 생성
ax[1, 1].imshow(Z)  # 네 번째 서브플롯에 이미지 그리기

plt.show()
```

```python
# 산점도 그리기

import matplotlib.pyplot as plt
import numpy as np

N = 30

x = np.random.rand(N)  # 0부터 1까지의 랜덤 숫자 30개 생성
y = np.random.rand(N)

colors = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)  # s: 점 크기, c: 색상, alpha: 투명도

plt.show()
```

```python
x = np.arange(3)
years = ["2010", "2011", "2012"]
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.bar(x, domestic)
plt.bar(x, foreign)
plt.xticks(x, years)

plt.show()
```

```python
x = np.arange(3)
years = ["2010", "2011", "2012"]
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.barh(x, domestic, height=0.25)
plt.barh(x - 0.3, foreign, height=0.25)
plt.yticks(x, years)

plt.show()
```
